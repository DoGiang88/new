```python
chieucao = int(input("Nhập chiều cao mặt tiền: "))
chieurong = int(input("Nhập chiều rộng mặt tiền: "))
window = int(input("Nhập cạnh của cửa sổ hình vuông: "))
tree = int(input("Nhập độ lớn của cây: "))
```

```python
import turtle
p = turtle.Turtle ()

#draw a house
#draw front face
p.penup ()
p.goto(-150,50)
p.pendown ()
p.pensize (5)
p.fd (chieurong)
p.rt (90)
p.fd (chieucao)
p.rt (90)
p.fd (chieurong)
p.rt (90)
p.fd (chieucao)

#draw roof
##draw triangle
p.rt (150)
p.begin_fill ()
p.pencolor ("#900C3F")
p.fillcolor ("#FF5733")
p.circle (60,steps = 3)
##draw rectangle
p.penup ()
p.goto (-45,50)
p.left (60)
p.pendown ()
p.fd (195)
p.left (90)
p.fd (60)
p.left(30)
p.fd (35)
p.left (60)
p.fd (230)
p.end_fill ()


#draw window
p.penup ()
p.goto (-100,30)
p.pensize (2)
p.pendown ()
p.begin_fill ()
p.pencolor ("#133392")
p.fillcolor ("#B2EBF3")
p.circle (window, steps =4)
p.end_fill ()

#draw main door
p.penup ()
p.goto (-45,-50)
p.pendown ()
p.begin_fill ()
p.rt (90)
p.fd (50)
p.rt (90)
p.fd (195)
p.rt (90)
p.fd (50)
p.end_fill ()

#draw chimney
p.penup ()
p.goto (-145,65)
p.pendown ()
p.rt (180)
p.begin_fill ()
p.pencolor ("black")
p.fillcolor ("#CACECE")
p.fd (70)
p.left (90)
p.fd (8)
p.rt (90)
p.fd (8)
p.rt (90)
p.fd (30)
p.rt (90)
p.fd (8)
p.rt (90)
p.fd (8)
p.left (90)
p.fd (45)
p.end_fill ()

#draw a cloud
p.penup ()
p.goto (200,250)
p.pendown ()
p.begin_fill ()
p.pensize (5)
p.pencolor ("#0854DB")
p.circle (30,50)
p.rt (30)
p.circle (30,60)
p.left (30)
p.circle (50,100)
```
