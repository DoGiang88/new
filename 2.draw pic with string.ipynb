# draw a house with tree

```python
chieucao = int(input("Nhập chiều cao mặt tiền: "))
chieurong = chieucao * 3
window = chieucao/5
tree = window*2
```

```python
import turtle
p = turtle.Turtle ()

#draw a house
#draw front face
p.penup ()
p.goto(-150,50)
p.pendown ()
p.pensize (5)
p.fd (chieurong)
p.rt (90)
p.fd (chieucao)
p.rt (90)
p.fd (chieurong)
p.rt (90)
p.fd (chieucao)

#draw roof
##draw triangle
p.rt (150)
p.begin_fill ()
p.pencolor ("#900C3F")
p.fillcolor ("#FF5733")
p.circle (chieucao*0.6,steps = 3)
##draw rectangle
p.penup ()
p.setpos (-45,50)
p.left (60)
p.pendown ()
p.fd (195)
p.left (90)
p.fd (60)
p.left(30)
p.fd (35)
p.left (60)
p.fd (230)
p.end_fill ()


#draw window
p.penup ()
p.goto (-100,30)
p.pensize (2)
p.pendown ()
p.begin_fill ()
p.pencolor ("#133392")
p.fillcolor ("#B2EBF3")
p.circle (window, steps =4)
p.end_fill ()

#draw main door
p.penup ()
p.goto (-45,-50)
p.pendown ()
p.begin_fill ()
p.rt (90)
p.fd (50)
p.rt (90)
p.fd (195)
p.rt (90)
p.fd (50)
p.end_fill ()

#draw chimney
p.penup ()
p.goto (-145,65)
p.pendown ()
p.rt (180)
p.begin_fill ()
p.pencolor ("black")
p.fillcolor ("#CACECE")
p.fd (70)
p.left (90)
p.fd (8)
p.rt (90)
p.fd (8)
p.rt (90)
p.fd (30)
p.rt (90)
p.fd (8)
p.rt (90)
p.fd (8)
p.left (90)
p.fd (45)
p.end_fill ()

#draw a tree
##tầng 1
p.penup ()
p.setpos (250,0)
p.left (90)
p.pendown ()
p.pensize (4)
p.begin_fill ()
p.pencolor ("#04390C")
p.fillcolor ("#77C183")
p.circle (-tree, steps=3)
p.end_fill ()
##tầng 2
p.penup ()
p.rt (90)
p.fd (tree*2/3)
p.pendown ()
p.begin_fill ()
p.left (90)
p.circle (-tree*25/20, steps=3)
p.end_fill ()
##tầng 3
p.penup ()
p.rt (90)
p.fd ((tree*25/20)*2/3)
p.pendown ()
p.begin_fill ()
p.left (90)
p.circle (-tree*3/2, steps=3)
p.end_fill ()
##thân cây
p.penup ()
p.rt (90)
p.fd ((tree*3/2)*2/3)
p.rt (15)
p.fd ((tree*3/2)*5/6)
p.pendown ()
p.pensize (5)
p.begin_fill ()
p.pencolor ("#361A09")
p.fillcolor ("#472612")
p.fd (tree*2)
p.left (105)
p.fd ((tree*3/2))
p.left (100)
p.fd (tree*2)
p.end_fill ()


#draw a tree
##tầng 1
p.penup ()
p.setpos (-250,0)
p.left (90)
p.pendown ()
p.pensize (4)
p.begin_fill ()
p.pencolor ("#04390C")
p.fillcolor ("#77C183")
p.circle (tree, steps=3)
p.end_fill ()
##tầng 2
p.penup ()
p.rt (90)
p.fd (tree*2/3)
p.pendown ()
p.begin_fill ()
p.left (90)
p.circle (tree*25/20, steps=3)
p.end_fill ()
##tầng 3
p.penup ()
p.rt (90)
p.fd ((tree*25/20)*2/3)
p.pendown ()
p.begin_fill ()
p.left (90)
p.circle (tree*3/2, steps=3)
p.end_fill ()
##thân cây
p.penup ()
p.rt (90)
p.fd ((tree*3/2)*2/3)
p.rt (15)
p.fd ((tree*3/2)*5/6)
p.pendown ()
p.pensize (5)
p.begin_fill ()
p.pencolor ("#361A09")
p.fillcolor ("#472612")
p.fd (tree*2)
p.left (105)
p.fd ((tree*3/2))
p.left (100)
p.fd (tree*2)
p.end_fill ()
```
